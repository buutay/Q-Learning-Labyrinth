```python
import math
import random
import numpy as np

M = [-1000, -1000, -1000, -1000, -1000, -1000, -1000, -1000, -1, -1, -1, -1000, -1, -1000, -1000, -1000, -1000, -1, -1000, -1, -1000, -1000, -1, -1, -1, -1, -1, -1000, -1000, -1000, -1, -1000, -1000, -1000, -1000, -1000, -1, -1, -1, -1, 1000, -1000, -1000, -1000, -1000, -1000, -1000, -1000, -1000]

def Action(n , i):
    k = 0
    if n == 0:
        if i in [0, 1, 2, 3, 4, 5, 6]:
            k = i
        else: k = i-7
    elif n == 1:
        if i in [42,43,44,45,46,47,48]:
            k = i
        else: k = i+7
    elif n == 2:
        if i%7 == 6:
            k = i
        else: k = i+1
    elif n == 3:
        if i%7 == 0:
            k = i
        else: k = i-1
    return k

    
A = [0,1,2,3]
P = [[0,0,0,0] for i in range(0,49)]
eps = 0.1
gamma = 0.9
alpha = 0.1
num_episode = 1000
it = 0
W = []
#for episode in range(100):
S0 = 8
while it < 200:
        xi = random.uniform(0,1)
        if xi > eps:
            action = A[np.argmax(P[S0])]
        else: 
            action = random.choice([0,1,2,3])
        
        S1 = Action(action, S0)
        
        R = M[S1]
        next_action = A[np.argmax(P[S1])]
        
        P[S0][action] += alpha*(R + gamma*P[S1][next_action] - P[S0][action])
        S0 = S1
        W.append(S0)
        it = it + 1
W
```
